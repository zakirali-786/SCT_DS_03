```python
import pandas as pd
```

```python
df = pd.read_excel(r"bank.xlsx")
```

```python
df.columns
```

```python
df['poutcome'].unique()
```

```python
df.select_dtypes(include="object")
```

```python
# seperating categorical columns....

cat_col = df[['job','marital','education','default','housing','loan','contact','month','poutcome']]
```

```python
# applying encoding on categorical columns....

from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

for col in cat_col:
    df[col] = encode.fit_transform(df[col].astype(str))
```

```python
df.info()
```

```python
col = df['y']
```

```python
# encoding output column of data....

df['y'] = pd.DataFrame(encode.fit_transform(col))

df
```

```python
x = df.drop(columns="y", axis =1)
y = df['y']
```

```python
# appliying scalling on dataframe....

from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

scale_df = scale.fit_transform(x)

x = pd.DataFrame(scale_df,columns=x.columns)

x
```

```python
# splitting data into training and testing....

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=786)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(min_samples_split=15)

model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)*100
```

```python
model.score(x_test,y_test)*100
```

```python
y_pred = model.predict(x_test)
```

```python
# importing metrix libraries for checking accuracy....

from sklearn.metrics import confusion_matrix,f1_score,accuracy_score,r2_score,classification_report
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
f1_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
